## Q_E	
25%까지 : 짧다, 25-75 : 보통, 75 이상 : 길다

```python
aelist = ['QaE', 'QbE', 'QcE', 'QdE', 'QeE', 'QfE', 'QgE', 'QhE', 'QiE', 'QjE', 'QkE', 'QlE', 'QmE', 'QnE', 'QoE', 'QpE', 'QqE', 'QrE', 'QsE', 'QtE']
```

```python
# 1분위 값과 3분위값 확인
for i in aelist:
    print(i ,'25%', ':', train[i].quantile(.25))
    print(i ,'75%', ':', train[i].quantile(.75))
    print('*'*30)
```

```python
# 1분위값 이하는 Small, 3분위 이상은 Big, 나머지는 Middle
for i in aelist:
    train[i+'_new'] = np.where(train[i]<=train[i].quantile(.25),'Small',train[i])
    train[i+'_new'] = np.where(train[i]>=train[i].quantile(.75),'Big',train[i+'_new'])
    train[i+'_new'] = np.where((train[i]>=train[i].quantile(.25)) & (train[i]<=train[i].quantile(.75)),'Middle',train[i+'_new'])
   
```

```python
sns.countplot(x="QaE_new", data=train)
```

## wf_01,02,03
셋 중 하나라도 1이 있는 행 -> 컬럼하나 더 생성 1,
셋 다 0 -> 0

```python
df = train.loc[:,['wf_01','wf_02','wf_03']]
df_list= df.values.tolist()
```

```python
wf_new = []
for i in df_list:
    if i == [0,0,0] :
        wf_new.append(0)
    else :
        wf_new.append(1)
        
len(wf_new)        
```

## wr
'wr_03','wr_06','wr_09','wr_11' : 0 > 1 이 중 2개 이상 1이라고 답했냐 아니냐로 분류 후 (컬럼하나 더 생성 , 1: 있다, 0 :없다)

```python
dr = train.loc[:,['wr_03','wr_06','wr_09','wr_11']]
dr.head()
```

```python
dr_list= dr.values.tolist()
```

```python
wr_new = []
for i in dr_list:
    if i.count(1) >= 2 :
        wr_new.append(1)
    else :
        wr_new.append(0)
        
len(wr_new)        
```

## tp__(01~10)
0,1 그룹, 2-5 그룹, 6-7 그룹

```python
def Filter(x):
    if x<2:
        return 0
    elif x<6:
        return 1
    else:
        return 2    
    
tp_list=[col for col in train.columns if 'tp' in col]

def get_tp(df):
    for col in tp_list:
        df[col+'_new']=df[col].apply(Filter)
    return df

train=get_tp(train)
```

```python
tp_new_list = [col+'_new' for col in tp_list]
train[tp_new_list].head()
```

####  tp의 전반적성향 파악해서 하나의 컬럼으로 만드는 방법 생각하기

#### tp 대표 성향 뽑아내기

```python
for i in range(5):
    train['tp'+str(i+1).zfill(2)+'-'+'tp'+str(i+6).zfill(2)] = train['tp'+str(i+1).zfill(2)] - train['tp'+str(i+6).zfill(2)]


tp_diff_list=[col for col in train.columns if 'tp' and '-' in col]
tp_diff = train[tp_diff_list]
tps = abs(tp_diff).idxmax(axis=1)
tp_rep = []
for i in range(len(train)):
    if tps[i] == 'tp01-tp06':
        if train[tps[i][0:4]][i] > train[tps[i][5:9]][i]:
            tp_rep.append(tps[i][0:4])
        else:
            tp_rep.append(tps[i][5:9])
            
    elif tps[i] == 'tp02-tp07':
        if train[tps[i][0:4]][i] > train[tps[i][5:9]][i]:
            tp_rep.append(tps[i][0:4])
        else:
            tp_rep.append(tps[i][5:9])
            
    elif tps[i] == 'tp03-tp08':
        if train[tps[i][0:4]][i] > train[tps[i][5:9]][i]:
            tp_rep.append(tps[i][0:4])
        else:
            tp_rep.append(tps[i][5:9])
            
    elif tps[i] == 'tp04-tp09':
        if train[tps[i][0:4]][i] > train[tps[i][5:9]][i]:
            tp_rep.append(tps[i][0:4])
        else:
            tp_rep.append(tps[i][5:9])
            
    else:
        if train[tps[i][0:4]][i] > train[tps[i][5:9]][i]:
            tp_rep.append(tps[i][0:4])
        else:
            tp_rep.append(tps[i][5:9])

train['tp'] = tp_rep
train[['tp']]
```

```python
# 각 요소당 전체 합 구하기
S_sum=[]
M_sum=[]
B_sum=[]

for i in QE_list :
#     S = pd.value_counts(QE[i].values)["Small"]
    S_sum.append(pd.value_counts(QE[i].values)["Small"])
    M_sum.append(pd.value_counts(QE[i].values)["Middle"])
    B_sum.append(pd.value_counts(QE[i].values)["Big"])
print("S_sum", S_sum)
print("M_sum",M_sum)
print("B_sum",B_sum)
    
```

```python
# 각 요소당 전체 평균 구하기
S_mean = sum(S_sum)/ 45532
print("S_mean", S_mean)
M_mean = sum(M_sum)/ 45532
print("M_mean", M_mean)
B_mean = sum(B_sum)/ 45532
print("B_mean", B_mean)
```

```python
#QE의 모든 행을 list화
QE_rowlist= QE.values.tolist()
```

```python
QEs_above = []

for i in range(0,len(train)):
    if QE_rowlist[i].count('Small') > S_mean :
        QEs_above.append(1)
    else : 
        QEs_above.append(0)
len(QEs_above)    
```

```python
QEm_above = []

for i in range(0,len(train)):
    if QE_rowlist[i].count('Middle') > M_mean :
        QEm_above.append(1)
    else : 
        QEm_above.append(0)
len(QEm_above)    
```

```python
QEb_above = []

for i in range(0,len(train)):
    if QE_rowlist[i].count('Big') > B_mean :
        QEb_above.append(1)
    else : 
        QEb_above.append(0)
len(QEb_above) 
```

```python
QE_above =[]
for i in range(0,len(train)):
    above = QEs_above[i]+QEm_above[i]+QEb_above[i]
    QE_above.append(above)
print(len(QE_above))
QE_above.count(2)
```

### other_cnt 변수 추가

```python
def score(col):
    score = 0
    
    if col==0:
        score+=1
    
    if col=='Other':
        score+=1
        
    return score
```

```python
train2 = train[["education", "engnat", "hand", "married", "race", "religion", "urban"]]
train2['other_cnt'] = train2['education'].apply(score) + train2['engnat'].apply(score) + train2['hand'].apply(score) + train2['married'].apply(score) +  train2['race'].apply(score) + train2['religion'].apply(score) + train2['urban'].apply(score)
```
