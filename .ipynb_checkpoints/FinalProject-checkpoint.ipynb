## QaA 막대 그래프 이용해 이상값 확인하기

```python
Q_list = list(train)[:40]
Q_list

Q_A_list = []
for i in range(len(Q_list)):
    if i % 2 ==0 :
        Q_A_list.append(Q_list[i])
        
Q_A_list        
```

```python
pd.DataFrame(train[Q_A_list[0]].value_counts())
```

```python
#sns 패키지 이용
sns.countplot(x="QaA", data=train)
```

```python
#group by 이용 count
train.groupby(Q_list[0])[Q_list[0]].count()
```

```python
#plot. 활용
train.groupby(Q_list[0])[Q_list[0]].count().plot.bar()
```

```python
#Q_A 한번에 출력

fig, ax = plt.subplots(20,1, figsize=(8,150))

for i in range(len(Q_A_list)):
    ax[i].bar(pd.DataFrame(train[Q_A_list[i]].value_counts()).index,pd.DataFrame(train[Q_A_list[i]].value_counts())[Q_A_list[i]] )
    ax[i].set_xlabel(Q_A_list[i])
```

#### QdA   응답1이 압도적으로 많고 5가 압도적으로 적음
#### QlA 응답5 이 압도적으로 많고 1,2가 압도적으로 적음
#### QqA 전체적으로 고름
#### 뭘 찾아야할지 모르겠음~!! 이상하다의 기준이 무엇인지 논의 필요!
